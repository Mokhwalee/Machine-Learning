## üßÆ üìå Summary Table of Key Formulas
| Metric                   | Formula                                      | When Used            | Output Meaning                |
|--------------------------|----------------------------------------------|----------------------|------------------------------|
| **Gini Impurity**        | Gini = $1 - Œ£ (p_i)^2   $                     | Classification       | 0 = pure, higher = mixed     |
| **Entropy**              | Entropy = $- Œ£ p_i log‚ÇÇ(p_i)   $               | Classification       | 0 = pure, higher = uncertainty |
| **Information Gain**     | IG = Entropy(parent) - Weighted Average Entropy(children) | Classification (best split) | Higher = better split |
| **Mean Squared Error (MSE)** | MSE = $(1 / N) Œ£ (y_i - ≈∑_i)^2      $     | Regression           | Lower = better predictions   |

## üìù Why Use Gini, Entropy, and MSE?
These metrics help measure impurity (for classification) or error (for regression), allowing decision trees to choose the best splits for making accurate predictions.

### üéØ Main Goal of Decision Trees:
- Split data into groups that are as pure as possible.
- Purity means nodes contain mostly or only one class (for classification) or have minimal prediction error (for regression).

### üü¢ Gini Impurity:
- Measures how often a randomly chosen element would be misclassified.
- Faster to compute than entropy (no logarithms).

### üü† Entropy & Information Gain:
- Entropy measures the level of disorder.
- Information Gain measures how much uncertainty is reduced after splitting.

### üü° Mean Squared Error (MSE):
- Measures prediction error for regression tasks.
- Penalizes larger errors more due to squaring.

## üü¢ Gini Impurity Example (Step-by-Step)
Assume we have 10 samples:
- Class A: 4 samples
- Class B: 6 samples

Gini Impurity Formula:
```text
Gini = 1 - Œ£ (p_i)^2
```
Where p_i is the proportion of samples of class i.

````python
total_samples = 10
count_A = 4
count_B = 6
p_A = count_A / total_samples
p_B = count_B / total_samples
gini = 1 - (p_A ** 2) - (p_B ** 2)
print(f"Gini Impurity = {gini:.2f}")
````

## üü† Entropy Example (Step-by-Step)
Entropy Formula:
```text
Entropy = - Œ£ p_i * log‚ÇÇ(p_i)
```
Higher entropy means more mixed or uncertain classes.

````python
import numpy as np
entropy = - (p_A * np.log2(p_A)) - (p_B * np.log2(p_B))
print(f"Entropy = {entropy:.3f}")
````

## üü° Mean Squared Error (MSE) Example (Step-by-Step)
MSE Formula:
```text
MSE = (1 / N) * Œ£ (y_i - ≈∑_i)^2
```
Used in regression tasks to minimize prediction error.

````python
y_actual = np.array([3, 4, 5])
y_predicted = np.array([2.5, 4.2, 5.1])
squared_errors = (y_actual - y_predicted) ** 2
mse = np.mean(squared_errors)
print("Squared Errors:", squared_errors)
print(f"Mean Squared Error (MSE): {mse:.2f}")
````

## üü¢ 1. Gini Impurity Example (Classification)

### Scenario:
Suppose we have **10 samples** with two classes:

| Class | Count |
|-------|-------|
| A     | 4     |
| B     | 6     |

### üßÆ Step-by-Step Calculation:

#### Step 1: Compute class probabilities.
```python
# Total number of samples
total_samples = 10

# Number of samples for each class
count_A = 4
count_B = 6

# Probability of each class
p_A = count_A / total_samples  # 4/10 = 0.4
p_B = count_B / total_samples  # 6/10 = 0.6
print("p_A =", p_A, ", p_B =", p_B)
```
**Output:**
```
p_A = 0.4 , p_B = 0.6
```

#### Step 2: Plug into Gini formula.
```python
# Gini Impurity formula
gini = 1 - (p_A ** 2) - (p_B ** 2)
print("Gini Impurity =", gini)
```
**Manual calculation:**
```
Gini = 1 - (0.4)^2 - (0.6)^2
     = 1 - 0.16 - 0.36
     = 1 - 0.52
     = 0.48
```
**Output:**
```
Gini Impurity = 0.48
```

## üü† 2. Entropy Example (Classification)

### üßÆ Step-by-Step Calculation:

#### Step 1: Use the same class probabilities.
```python
import numpy as np

entropy = - (p_A * np.log2(p_A)) - (p_B * np.log2(p_B))

```
**Manual calculation:**

### üîé Interpretation:

## üü° 3. Mean Squared Error (MSE) Example (Regression)

### Scenario:
| Actual (y) | Predicted (≈∑) |
|------------|---------------|
| 3          | 2.5           |
| 4          | 4.2           |
| 5          | 5.1           |

### üßÆ Step-by-Step Calculation:

#### Step 1: Compute squared errors for each data point.
```python
# Actual values
y_actual = np.array([3, 4, 5])

# Predicted values
y_predicted = np.array([2.5, 4.2, 5.1])

# Squared differences
squared_errors = (y_actual - y_predicted) ** 2
print("Squared Errors:", squared_errors)
```
**Output:**
```
Squared Errors: [0.25 0.04 0.01]
```

#### Step 2: Compute MSE (mean of squared errors).
```python
mse = np.mean(squared_errors)
print("Mean Squared Error (MSE):", mse)
```
**Manual calculation:**
```
MSE = (0.25 + 0.04 + 0.01) / 3
    = 0.30 / 3
    = 0.10
```
**Output:**
```
Mean Squared Error (MSE): 0.1
```

### ‚úÖ Interpretation:
- **Lower MSE:** Predictions are close to actual values (better fit).
- **Higher MSE:** Predictions are far from actual values (worse fit).

## üèóÔ∏è Decision Tree Classifier Example (Iris Dataset)
Let's apply the classification concepts in Python.

````python
# üå≥ Decision Tree Classifier Example

# üì• 1. Import Necessary Libraries
from sklearn.datasets import load_iris                             # Load the Iris dataset (built-in)
from sklearn.tree import DecisionTreeClassifier, plot_tree         # Decision Tree classifier and visualization tool
import matplotlib.pyplot as plt                                     # For plotting the decision tree
from sklearn.model_selection import train_test_split               # For splitting data into train/test sets
from sklearn.metrics import accuracy_score                         # To evaluate model performance


# üå∏ 2. Load the Iris Dataset
# Iris dataset contains 150 samples of 3 flower species
iris = load_iris()

# X = input features (sepal length, sepal width, petal length, petal width)
X = iris.data

# y = target labels (0 = setosa, 1 = versicolor, 2 = virginica)
y = iris.target


# ‚úÇÔ∏è 3. Split Data into Train and Test Sets
# 70% training data, 30% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# üå≥ 4. Initialize the Decision Tree Classifier
# Using Gini impurity as the splitting criterion
# Limiting maximum depth to 3 levels to avoid overfitting
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)


# üèãÔ∏è‚Äç‚ôÇÔ∏è 5. Train (Fit) the Model on Training Data
clf.fit(X_train, y_train)  # The tree learns how to split based on the training data


# üîÆ 6. Make Predictions on Test Data
y_pred = clf.predict(X_test)  # Predict the class labels for test data


# üìè 7. Evaluate the Model Performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy on Test Set:", accuracy)


# üé® 8. Visualize the Decision Tree Structure
plt.figure(figsize=(12, 8))  # Set the size of the plot

# Plot the decision tree
plot_tree(clf,
          filled=True,                                  # Fill colors based on classes
          feature_names=iris.feature_names,             # Add feature names (for better readability)
          class_names=iris.target_names)                # Add class names (setosa, versicolor, virginica)

plt.title("Decision Tree Classifier on Iris Dataset")   # Add a title to the plot
plt.show()                                              # Display the plot
````

## üèóÔ∏è Decision Tree Regressor Example (Synthetic Data)
Let's apply the regression concepts in Python.

````python
# üå≥ Decision Tree Regressor Example
# ============================
# üì• 1. Import Necessary Libraries
# ============================
from sklearn.tree import DecisionTreeRegressor                      # Decision Tree for regression tasks
import numpy as np                                                  # For numerical operations and generating synthetic data
import matplotlib.pyplot as plt                                     # For plotting the results
from sklearn.metrics import mean_squared_error                      # To evaluate the model performance (error measurement)

# üé≤ 2. Generate Synthetic Nonlinear Data (Sine Wave with Noise)
# Generate 80 random X points between 0 and 5, sorted for visualization clarity
X = np.sort(5 * np.random.rand(80, 1), axis=0)

# Create target values (y) as a sine wave with added random noise (to simulate real-world noisy data)
y = np.sin(X).ravel() + np.random.randn(80) * 0.1  # .ravel() flattens the array to 1D


# üå≥ 3. Initialize the Decision Tree Regressor
# max_depth=3 limits how deep the tree can grow (to avoid overfitting)
regressor = DecisionTreeRegressor(max_depth=3)


# üèãÔ∏è‚Äç‚ôÇÔ∏è 4. Train (Fit) the Model on the Data
regressor.fit(X, y)  # The regressor learns the best splits to approximate the sine wave


# üîÆ 5. Generate Test Points and Make Predictions
# Create test points for smooth curve plotting (from 0.0 to 5.0 with small steps)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]  # Reshape for compatibility

# Predict using the trained regressor
y_pred = regressor.predict(X_test)


# üé® 6. Visualize the Regression Results
plt.figure(figsize=(10, 6))                                         # Set plot size

# Scatter plot of the original training data (blue dots)
plt.scatter(X, y, s=20, label="Training Data")

# Line plot of the predictions (red line shows the piecewise-constant predictions)
plt.plot(X_test, y_pred, color="red", label="Prediction")

# Add labels and title
plt.xlabel("X")
plt.ylabel("y")
plt.title("Decision Tree Regression Example")
plt.legend()
plt.show()


# üìè 7. Evaluate the Model Performance (Mean Squared Error)
# Calculate MSE on the training data (how well the regressor fits the training set)
mse = mean_squared_error(y, regressor.predict(X))
print("Mean Squared Error (MSE):", mse)
````
